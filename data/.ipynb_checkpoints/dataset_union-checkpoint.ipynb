# BOOKSTORES

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')

combined_df = pd.concat([train_df, test_df])

output_path = 'Combined_Satisfaction.csv'
combined_df.to_csv(output_path, index=False)

combined_df.head(), output_path
```

```python
def plot_differences_with_values(train_info, test_info, differences):
    row_percentage_diff = (differences['row_difference'] / train_info['rows']) * 100
    
    plt.figure(figsize=(14, 7))
    
    bars = plt.bar(['Train', 'Test'], [train_info['rows'], test_info['rows']], color=['blue', 'orange'])
    plt.title('Number of Rows')
    plt.ylabel('Count')
    for bar, value in zip(bars, [train_info['rows'], test_info['rows']]):
        plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() - bar.get_height() * 0.1,
                 f'{value}', ha='center', color='white', weight='bold')
    
    plt.text(0.5, max(train_info['rows'], test_info['rows']) * 0.5,
             f'Difference: {differences["row_difference"]}\n({row_percentage_diff:.2f}%)', 
             ha='center', color='black', weight='bold')

    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(14, 7))
    
    bars = plt.bar(['Train', 'Test'], [train_info['columns'], test_info['columns']], color=['blue', 'orange'])
    plt.title('Number of Columns')
    plt.ylabel('Count')
    for bar, value in zip(bars, [train_info['columns'], test_info['columns']]):
        plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() - bar.get_height() * 0.1,
                 f'{value}\n100%', ha='center', color='white', weight='bold')

    plt.tight_layout()
    plt.show()

train_info = {
    "rows": train_df.shape[0],
    "columns": train_df.shape[1],
    "column_names": set(train_df.columns)
}

test_info = {
    "rows": test_df.shape[0],
    "columns": test_df.shape[1],
    "column_names": set(test_df.columns)
}

differences = {
    "row_difference": abs(train_info["rows"] - test_info["rows"]),
    "column_difference": abs(train_info["columns"] - test_info["columns"]),
    "common_columns": train_info["column_names"].intersection(test_info["column_names"]),
    "train_unique_columns": train_info["column_names"].difference(test_info["column_names"]),
    "test_unique_columns": test_info["column_names"].difference(train_info["column_names"])
}

plot_differences_with_values(train_info, test_info, differences)
```


Number of rows: Notable increase in data.

Number of columns: there was no change.
